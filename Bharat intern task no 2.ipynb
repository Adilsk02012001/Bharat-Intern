```python
import seaborn as sb
import matplotlib.pyplot as mpl
```

```python
sb.set(style = "blue", font = "Arial")
```

```python
_, ax = mpl.subplots(figsize = (10, 7))
sb.histplot(data = dataframe, x = "Age", hue = "Survived", multiple = "stack", kde = True, palette=["blue", "black"], ax = ax)
ax.set_title("Histogram for Age with Survival")
ax.set_xlabel("Age")
ax.set_ylabel("Number of passengers")
ax.legend(title = "Legends", labels = ["Survived", "Not survived"])
mpl.plot()
```

```python
embarked_counts = dataframe['Embarked'].value_counts()
mpl.figure(figsize=(8, 8))
mpl.pie(embarked_counts, labels=embarked_counts.index, autopct='%1.1f%%', colors=['#66b3ff', '#99ff99', '#ff9999'])
mpl.title("Distribution of Passengers by Embarked")
mpl.legend(["Q", "S", "C"])
mpl.show()
```

```python
_, ax = mpl.subplots(figsize = (10, 7))
sb.countplot(data = dataframe, x = "Embarked", hue = "Survived", palette=["black", "blue"], ax = ax)
ax.set_title("Countplot for Embarked with Survival")
ax.set_xlabel("Embarked")
ax.set_xticklabels(["Q", "S", "C"])
ax.set_ylabel("Number of passengers")
ax.legend(title = "Legends", labels = ["Not Survived", "Survived"])
mpl.plot()
```

```python
_, ax = mpl.subplots(figsize = (10, 7))
sb.histplot(data = dataframe, x = "Fare", hue = "Survived", multiple = "stack", kde = True, palette = ["green", "black"], ax = ax)
ax.set_title("Histogram for Fare with Survival")
ax.set_xlabel("Fare")
ax.set_ylabel("Number of passengers")
ax.legend(title = "Legends", labels = ["Survived", "Not survived"])
mpl.plot()
```

```python
ax = sb.countplot(data = dataframe, x="Survived", hue = "Survived", palette = ["yellow", "orange"])
ax.set_xlabel("Survived")
ax.set_ylabel("Count")
ax.set_title("Survival Count")
ax.legend(title = "Legends", labels = ["Not Survived", "Survived"])
mpl.xticks([0,1],["Male", "Female"])
mpl.show()
```
